```python
import pickle, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy import stats, integrate
from tqdm import tqdm_notebook
```

```python
def calculate_density(percentile, standard_errors, residuals):
    num_within_quantile = 0
    for se, resid in zip(standard_errors, residuals):
        norm = stats.norm(loc=0, scale=se)
        lower_bound = norm.ppf(0.5-percentile/2)
        upper_bound = norm.ppf(0.5+percentile/2)
        if lower_bound <= resid <= upper_bound:
            num_within_quantile += 1
    density = num_within_quantile / len(residuals)
    return density

predicted_pi = np.linspace(0, 1, 20)
```

## Making Calibration Curves for Ensembles with various dropouts

```python
import multiprocessing as mp

def observed_pi_calc(std, res, dropout):
    return [calculate_density(quantile, std, res)
            for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with %s%% dropout' % dropout)]
    
observed_pi_input_table = [[] for _ in range(7)]
observed_pi_input_table[0] = [ens_v_std, ens_v_res, 0]
observed_pi_input_table[1] = [ens_d05_std, ens_d05_res, 5]
observed_pi_input_table[2] = [ens_d10_std, ens_d10_res, 10]
observed_pi_input_table[3] = [ens_d15_std, ens_d15_res, 15]
observed_pi_input_table[4] = [ens_d_std, ens_d_res, 20]
observed_pi_input_table[5] = [ens_d25_std, ens_d25_res, 25]
observed_pi_input_table[6] = [ens_d30_std, ens_d30_res, 30]
                              
pool = mp.Pool(processes=7)
out  = pool.map(observed_pi_calc, observed_pi_input_table)

ens_v_observed_pi   = out[0]
ens_d05_observed_pi = out[1]
ens_d10_observed_pi = out[2]
ens_d15_observed_pi = out[3]
ens_d_observed_pi   = out[4]
ens_d25_observed_pi = out[5]
ens_d30_observed_pi = out[6]
```

```python
ens_d05_observed_pi = [calculate_density(quantile, ens_d05_std, ens_d05_res)
                       for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 05% dropout')]
ens_d10_observed_pi = [calculate_density(quantile, ens_d10_std, ens_d10_res)
                       for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 10% dropout')]
ens_d15_observed_pi = [calculate_density(quantile, ens_d15_std, ens_d15_res) 
                       for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 15% dropout')]
ens_d25_observed_pi = [calculate_density(quantile, ens_d25_std, ens_d25_res)
                       for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 25% dropout')]
ens_d30_observed_pi = [calculate_density(quantile, ens_d30_std, ens_d30_res)
                       for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 30% dropout')]
ens_d_observed_pi = [calculate_density(quantile, ens_d_std, ens_d_res)
                     for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with 20% dropout')]
ens_v_observed_pi = [calculate_density(quantile, ens_v_std, ens_v_res)
                     for quantile in tqdm_notebook(predicted_pi, desc='Calibration - Case with no dropout')]
```

```python
fig = plt.figure(figsize=(8, 8))

ax_ideal = sns.lineplot([0, 1], [0, 1], label='ideal')
_ = ax_ideal.lines[0].set_linestyle('--')
_ = ax_ideal.set_title("Adding dropout (up to 20%) to the CGCNN ensemble\n" +
                       "makes the calibration curve converge to the ideal")

ens_v_ax_gp = sns.lineplot(predicted_pi, ens_v_observed_pi, label='Ensemble - No dropout')
_ = ens_v_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_v_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_v_ax_gp.set_xlim([0, 1])
_ = ens_v_ax_gp.set_ylim([0, 1])
_ = ens_v_ax_gp.lines[1].set_linewidth(3)

ens_d05_ax_gp = sns.lineplot(predicted_pi, ens_d05_observed_pi, label='Ensemble - 5% dropout')
_ = ens_d05_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d05_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d05_ax_gp.set_xlim([0, 1])
_ = ens_d05_ax_gp.set_ylim([0, 1])

ens_d10_ax_gp = sns.lineplot(predicted_pi, ens_d10_observed_pi, label='Ensemble - 10% dropout')
_ = ens_d10_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d10_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d10_ax_gp.set_xlim([0, 1])
_ = ens_d10_ax_gp.set_ylim([0, 1])

ens_d15_ax_gp = sns.lineplot(predicted_pi, ens_d15_observed_pi, label='Ensemble - 15% dropout')
_ = ens_d15_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d15_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d15_ax_gp.set_xlim([0, 1])
_ = ens_d15_ax_gp.set_ylim([0, 1])

ens_d_ax_gp = sns.lineplot(predicted_pi, ens_d_observed_pi, label='Ensemble - 20% dropout')
_ = ens_d_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d_ax_gp.set_xlim([0, 1])
_ = ens_d_ax_gp.set_ylim([0, 1])

ens_d25_ax_gp = sns.lineplot(predicted_pi, ens_d25_observed_pi, label='Ensemble - 25% dropout')
_ = ens_d25_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d25_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d25_ax_gp.set_xlim([0, 1])
_ = ens_d25_ax_gp.set_ylim([0, 1])

ens_d30_ax_gp = sns.lineplot(predicted_pi, ens_d30_observed_pi, label='Ensemble - 30% dropout')
_ = ens_d30_ax_gp.set_xlabel('Expected prediction interval')
_ = ens_d30_ax_gp.set_ylabel('Observed prediction interval')
_ = ens_d30_ax_gp.set_xlim([0, 1])
_ = ens_d30_ax_gp.set_ylim([0, 1])

fig.savefig("CGCNN_ensemble_dropout_CCurve_v2.png")
```
